## 1. HSSR Encoding and Decoding

HSSR (Hierarchical Structure Sequence Representation) is our novel format for representing crystal structures as text sequences.

```python
# Initialize encoder
encoder = HSSREncoder()

# Sample MOF-5 features
mof5_features = {
    'structure_id': 'MOF-5',
    'formula': 'Zn4O(C8H4O4)3',
    'lattice': {
        'a': 25.832, 'b': 25.832, 'c': 25.832,
        'alpha': 90.0, 'beta': 90.0, 'gamma': 90.0
    },
    'space_group': 'Fm-3m',
    'space_group_number': 225,
    'crystal_system': 'cubic',
    'topology': 'pcu',
    'metal_node': 'Zn4O tetrahedral cluster',
    'linker': '1,4-benzenedicarboxylate',
    'pore_diameter': 11.0,
    'surface_area': 3800,
    'density': 0.59,
    'void_fraction': 0.78
}

# Encode to HSSR
hssr = encoder.encode(mof5_features, 'MOF')
print("Encoded HSSR:")
print(hssr)
```

```python
# Validate HSSR
is_valid, errors = encoder.validate(hssr)
print(f"Valid: {is_valid}")
if errors:
    print(f"Errors: {errors}")
```

```python
# Decode HSSR back to features
decoded = encoder.decode(hssr)
print("Decoded features:")
for key, value in decoded.items():
    print(f"  {key}: {value}")
```

## 2. Text Description Generation

Generate natural language descriptions at different levels of detail.

```python
# Initialize text generator
text_gen = TextGenerator()

# Generate Level 1 description (formula-based)
level1 = text_gen.generate(mof5_features, level=1)
print("Level 1 Description:")
print(level1)
```

```python
# Generate Level 2 description (structure-based)
level2 = text_gen.generate(mof5_features, level=2)
print("Level 2 Description:")
print(level2)
```

## 4. Evaluation Metrics

Compute evaluation metrics for model outputs.

```python
# Example S2T evaluation
predictions = [
    "MOF-5 is a zinc-based metal-organic framework with high porosity.",
    "HKUST-1 is a copper MOF with paddlewheel clusters."
]

references = [
    "MOF-5 is a zinc-based metal-organic framework with large pores.",
    "HKUST-1 is a copper-based MOF featuring paddlewheel secondary building units."
]

# Compute metrics
metrics = compute_metrics(predictions, references, task='s2t')
print_metrics(metrics, task='s2t')
```

```python
# Example T2S evaluation
hssr_predictions = [
    "[CRYSTAL] [META] MOF MOF-5 [COMP] Zn4O(C8H4O4)3 [/CRYSTAL]",
    "[CRYSTAL] [META] MOF HKUST-1 [COMP] Cu3(C9H3O6)2 [/CRYSTAL]"
]

hssr_references = [
    "[CRYSTAL] [META] MOF MOF-5 [COMP] Zn4O(C8H4O4)3 [/CRYSTAL]",
    "[CRYSTAL] [META] MOF HKUST-1 [COMP] Cu3(C9H3O6)2 [/CRYSTAL]"
]

# Compute metrics
t2s_metrics = compute_metrics(hssr_predictions, hssr_references, task='t2s')
print_metrics(t2s_metrics, task='t2s')
```
